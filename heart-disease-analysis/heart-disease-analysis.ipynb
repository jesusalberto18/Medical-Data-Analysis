# HEART DISEASE ANALYSIS

## In this notebook there is an analysis of a group of patients who have suffered a heart disease in their lifetime and the values of the signs they showed that indicate the early stages of a heart disease.

For this analysis we have to import 4 essencial libraries to use:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Now we'll upload the dataset that we'll working on: 'heart.csv'

```python
df = pd.read_csv('heart.csv')
```

And now we take a quick look to its info:

```python
df.info()
```

As we can see, there are not null values nor objects in this dataset, all of the cells are integers so we can proceed to make our analysis without any problem.

Take a look to the first and the last 10 rows of the dataset:

```python
df.head(10)
```

```python
df.tail(10)
```

Now we apply a quick and simple statistical analysis that lets us make our first conclusion about the info in the dataset:

```python
df.describe()
```

After checking the values in the statistical analysis performed before, we can see that there's a close relation between the chest pain and the possibility of enduring a heart disease. This relation is mostly seen by the angina produced by exercising. The harder the exercise, the deeper the pain; and the evolved the disease, the more accute the pain.

The next scatter plot shows the distribution of heart disease between the individuals seen  in this dataset:

```python
df.plot.scatter(x=3,y=4,title='Heart Disease Analysis Scatter Plot',subplots=False)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.show()
```

Now we're going to evaluate and perform an analysis to each column in the dataset.

Let's start with the first column that indicates the age of the individuals:

```python
print(df.groupby("age").size())
```

We can see here the range of ages among the individuals goes from 29 years old to 77 years old. In the plot below, we can see how the amount of individuals that share a similar or the same age can be found in the range between 40-60 years old, especially in their early and late 50's.

```python
sns.displot(df["age"], kde=True)
```

Now let's analyze the column that indicates the sex of the individuals:

```python
print(df.groupby("sex").size())
```

We identify sex with a 1=male & 0=female. We can see that in this dataset there are 1025 individuals split in 713 males and 312 females.

```python
sns.countplot(df["sex"])
```

Now we proceed to analyze the column that indicates the presence of chest pain felt by each individual:

```python
print(df.groupby("chest_pain").size())
```

We can see that over half individuals have presented at least one chest pain that anounces the development of a heart disease, but surprisingly about 49% of individuals who endure a heart disease have not had a chest pain that could be taken as a sign of a heart disease. 

```python
sns.displot(df["chest_pain"], kde=True)
```

The next step is to analyze the column that indicates the resting blood type of each individual:

```python
print(df.groupby("resting_blood_pressure (mm/Hg)").size())
```

A normal and healthy resting blood pressure is 120 mm/Hg or less, higher levels are considered hypertension. A hypertension crisis will show up when the resting blood pressure is above 180 mm/Hg and it's a very critical condition.

```python
sns.displot(df["resting_blood_pressure (mm/Hg)"], kde=True)
```

The next step is to analyze the column that indicates the levels of cholesterol found in the serum of each individual:

```python
print(df.groupby("serum_cholesterol (mg/dl)").size())
```

The serum cholesterol is a clear indicator of a future heart disease, it should be taken in the prevention and be part of any public health campaign. We can see that in this dataset it ranges from 126 mg/dl to 564 mg/dl, the range for optimal serum cholesterol is 125-200 mg/dl.

```python
sns.displot(df["serum_cholesterol (mg/dl)"], kde=True)
```

Now we have to analyze the column that indicates the blood sugar levels during the fasting stage of each individual:

```python
print(df.groupby("fasting_blood_sugar").size())
```

The fasting blood sugar is measured by 1=true & 0=false if the levels are equal to 120 mg/dl.

```python
sns.countplot(df["fasting_blood_sugar"])
```

The next step is to analyze the column that indicates the electrocardiographic results during the resting stage of each individual:

```python
print(df.groupby("resting_electrocardiographic_results").size())
```

```python
sns.displot(df["resting_electrocardiographic_results"], kde=True)
```

Now we proceed to analyze the column that indicates the maximum heart rate that each individual achieves, this is frequently achieved phisiologically during an exercise:

```python
print(df.groupby("max_heart_rate_achieved").size())
```

The heart rate is closely related to the blood pressure, so it can be considered as having the same values for each variable. That means that a normal heart rate is somewhere close to 120 beats per minute.

```python
sns.displot(df["max_heart_rate_achieved"], kde=True)
```

Finally we proceed to analyze the column that indicates the presence of angina induced by any level of physical activity:

```python
print(df.groupby("exercise_induced_angina").size())
```

The exercise induced angina is measured by 1=yes & 0=no. Like in the case of chest pain, it's surprising that many of the individuals that have endured a heart disease have not felt any pain at all; but in this case 2/3 of individuals have not had an angina induced by exercise, so it can't always be taken as a sign that there won't be a heart disease developing.

```python
sns.countplot(df["exercise_induced_angina"])
```

We can get to the conclusion that the most important variables for predicting if someone has already developed a heart disease are: Age & Chest Pain.

Also, we can get to the conclusion that the most important variables for predicting if someone is going to develop a heart disease in the future are: Sex & Serum Cholesterol.

In the next plot we can see the relation between the age of each individual and the presence of chest pain:

```python
sns.displot(x="age", y="chest_pain", data=df)
```

As we can see here, there's a higher presence of chest pain that indicates a heart disease in those individuals between 40-60 years old. Yet around in half of cases we can see that the chest pain is not present.

In the next plot we can see the relation between the age of each individual and the levels of serum cholesterol:

```python
sns.displot(x="age", y="serum_cholesterol (mg/dl)", data=df)
```

We can notice that individuals who tend to have higher levels of serum cholesterol are over 30 years old, increasd¡ing its prevalence as an individual get older. 

In the next plot we can see the relation between the sex of each individual and the presence of chest pain:

```python
sns.displot(x="sex", y="chest_pain", data=df, kind="kde", rug=True)
```

We notice that men tend to have a more frequent presence of chest pain that indicates a heart disease than women, so does the absence of it which could be a sign of more silent heart diseases in men. 

In the next plot we can see the relation between the sex of each individual and the levels of serum cholesterol:

```python
sns.displot(x="sex", y="serum_cholesterol (mg/dl)", data=df, kind="kde", rug=True)
```

Just as in our last plot, we notice that men have higher levels of serum cholestrol compared to women. This means that it should be taken care of and be often reviewed since it could be a cause of a future heart disease.

DISCLAIMER: The evaluation of the relation among these 4 variables in this analysis does not mean that the other variables in this dataset are not important for getting to a conclusion about heart disease and should be considered in a medical study.

## IT CAN BE PREVENTED BY FOLLOWING A HEALTHY LIFESTYLE:
1. DON'T SMOKE OR USE TOBACCO. Even if you're not a smoker, try to avoid secondhand smoke. Cigarettes contain harmful chemicals and the smoke reduces the oxygen in the blood, incresing blood pressure levels and damaging the heart and blood vessels.
2. DO EXERCISE. It's recommended to practice 30-60 minutes of physical activity daily to keep your heart health, as it helps you reduce high blood pressure levels and high serum cholesterol levels.
3. EAT HEALTHY. Have a healthier diet that include fruits, vegetables and grains; also you should avoid the consumption or exceeding the intake of salt, sugar, processed carbohydrates, alcohol, trans fats and saturated fats.
4. CONTROL YOUR WEIGHT. Being overweight increases the risk of heart disease. The risk of heart disease is higher if the waist measurement is over 40 inches in men and over 35 inches in women, so we should aim to a routine that helps us lose weight and keep it in a regular and desirable condition, such as practicing some exercise or a healthy diet.
5. GET GOOD QUALITY SLEEP. People who don't get enough sleep have a higher risk of heart disease since it encourages obesity and high blood pressure.
6. MANAGE STRESS. Find a healthy way to cope with stress, such as meditation, physical activity or relaxation exercises; this way you can improve your heart health.
7. REGULAR HEALTH SCREENING. We've mentioned many times before that having high blood pressure and serum cholesterol can damage your heart, but the hardest part is that if you never have a health secreening you could never know whether you endure any of those. Once an individual has reached the age 18, they should make some medical exams for testing the blood pressure, the serum cholesterol levels and screening diabetes type 2.
